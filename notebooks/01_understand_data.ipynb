```python
# The Dataset has no duplicated values.
Data.duplicated().sum()
```

```python
# The Dataset have null values in (BasePay, Benefits, Notes, and Status) columns.
Data.isnull().sum()
```

```python
# This Charts show the null values in the dataset.
# barplot is used to show the count of null values in each column.
# And show the percentage of null values in each column.
plt.figure(figsize=(10, 6))
sns.barplot(x=Data.isnull().sum().index, y=Data.isnull().sum().values)
plt.title("Count of Null Values in Each Column")
plt.xlabel("Columns")
plt.ylabel("Count of Null Values")
plt.xticks(rotation=45)
plt.show()
```

```python
# where the percentage of null values in each column is calculated by dividing the count of null values by the total number of rows in the dataset and multiplying by 100.
null_percentage = (Data.isnull().sum() / len(Data)) * 100
plt.figure(figsize=(10, 6))
sns.barplot(x=null_percentage.index, y=null_percentage.values)
plt.title("Percentage of Null Values in Each Column")
plt.xlabel("Columns")
plt.ylabel("Percentage of Null Values")
plt.xticks(rotation=45)
plt.show()
```

###  3.2 Understanding a EmployeeName Column

```python
Data["EmployeeName"].value_counts().head(20)
```

```python
# This chart shows the top 10 Employee Names in the dataset.
plt.figure(figsize=(10, 6))
sns.barplot(x=Data["EmployeeName"].value_counts().head(10).index, y=Data["EmployeeName"].value_counts().head(10).values)
plt.title("Top 10 Employee Names in the Dataset")
plt.xlabel("Employee Names")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# This Chart showing the bottom 10 Employee Names in the dataset.
plt.figure(figsize=(10, 6))
sns.barplot(x=Data["EmployeeName"].value_counts().tail(10).index, y=Data["EmployeeName"].value_counts().tail(10).values)
plt.title("Bottom 10 Employee Names in the Dataset")
plt.xlabel("Employee Names")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```
